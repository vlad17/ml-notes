# Problem 9.10

_Prove Remark 9.4._

Generally, suppose we have an SCM on a DAG $\mcG$ over variables $X, Z$, and possibly some others, such that
$$
p(z|x) = p(z|\mathrm{do}(x)) = f(z, x)\,.
$$

By completeness of do-calculus, this implies that $X\independent Z$ in $\mcG_{\underline{X}}$, so any open paths between $X$ and $Z$ must be directed paths from $X$ to $Z$ (any open path must start $X\rightarrow Y \cdots Z$, with $Y$ possibly equal to $Z$, but then it must entirely consist of right arrows, else the path is blocked). This implies $X\independent Z$ in $\mcG_{\overline Z}$ as well.

Then, the subgraph $X\rightarrow Z$ is interventionally sufficient. For any intervention on $X$ we simply equip the arrow with the kernel $f$, and give $X$ its marginal $p(x)$ in the larger SCM.

For an intervention on $Z$, the subgraph behaves like the empty graph, with $Z\sim \mathrm{do}(Z)$, and $X$ its marginal again.

This matches the larger SCM's behavior for all interventions.

The above proof does not assume anything about the larger SCM's faithfulness or structure, and thus applies to Remark 9.4's cases (i, ii, iii)

# Problem 9.12

_Given an SCM with roots $H,Z$ and $X\gets f(Z)+g(H)+N_X,Y\gets \alpha X+j(H) +N_Y$ with independent noises. How would you identify $\alpha$?._

Given an observational distribution $(H, Z, X, Y)$, we can perform least squares regression minimizing
$$
q_*=\argmin_q\E\ha{(X-q(Z))^2}\,\,.
$$
Since conditional expectation is $L^2$ projection, we find that this identifies a function such that
$$
q_*(Z)=\CE{X}{Z}=f(Z)\,\,,
$$
since $H,N_X\independent Z$. Then, since
$$
Y=\alpha X+j(H)+N_Y=\alpha(f(Z)+g(H)+N_X)+j(H)+N_Y\,\,,
$$
a corresponding second-stage regression
$$
p_*=\argmin_p\E\ha{(Y-p(q_*(Z)))^2}\,\,,
$$
finds the solution
$$
p_*(q_*(Z))=\CE{Y}{q_*(Z)}=\alpha q_*(Z)\,\,,
$$
since $q_*(Z)=f(Z)$ so $\CE{\alpha f(Z)}{q_*(Z)}=\alpha q_*(Z)$ while all other variables $H, N_X,N_Y$ are independent of $q_*(Z)$ because they're independent of $Z$ (and independence is maintained through the function $q_*$ thanks to the tower property).

This shows that the final least squares operation, a linear regression, recovers $\alpha$.

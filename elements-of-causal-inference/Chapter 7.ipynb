# Problem 7.12

_Allowing for zero coefficients, prove that two linear Gaussian SCMs $\mcG,\mcH$ are distribution equivalent iff they are Markov equivalent_.

$(\implies)$. If $\mcG,\mcH$ are distributionally equivalent, then the undirected skeleton must match.

Of course, for any implied joints to match at all, the variables must be the same.

Second, for any edge $X\rightarrow Y$ in $\mcG$ and choose weight $1$ and set the rest to $0$. This means $(X,Y)$ are independent of all other nodes in the graph. $\mcH$'s matching parameters for all other edges must then be 0, or $\mcH$ must not have any other edges. However, to maintain $\cov(X,Y)=1$, $\mcH$ must have an edge in some direction between $X,Y$ with weight 1. Applying this logic in the other direction shows our edges match undirectedly.

It remains to show that the v-structures match as well. Wlog we show one direction. Again for $X\rightarrow Y \leftarrow Z$ we in $\mcG$ we set those arrows to 1 and the rest to $0$. The only chain $X,Y,Z$ that can allow $\cov(X,Y)=1,\cov(Y,Z)=1$ but $X\independent Z$ is a collider at $Y$.

By Lemma 6.25, this shows Markov equivalence.

$(\impliedby)$. By Markov equivalence, $\mcG,\mcH$ share the same set of joint distributions. All joint distributions Markovian with respect to a linear Gaussian SCM must come from some parameterization of that SCM (since it's a parameterized model). As such, this direction is trivial.

By the factorization definition of Markovian (Gaussian has strictly positive, continuous density so we can apply the equivalence theorem here), $P$ must factorize into the parental structure of $\mcG$.

This factorization must be made from linear conditional Gaussian terms, since each node $X$ has a linear conditional Gaussian distribution given its parents (as it's jointly Gaussian with them).

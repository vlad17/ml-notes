# Problem 5.1

_Let $X$ be a $k$-Gaussian mixture with distinct peaks at $s_i$ and set $Y\gets X + N$ with $N\sim N(0,\sigma_N^2)$._

_(A) Why does $p(x|y)$ intuitively contain information about peaks, showing that the effect to cause mechanism is not independent of the marginal for the effect?_

First, in the true generative model, consider a multinomial $Z$ which determines which Gaussian $X$ is drawn from. This recovers the same mixture, with SCM with graph $Z\rightarrow X\rightarrow Y$.

Suppose the peaks are sufficiently well spaced relative to $\sigma_N^2$ such that for $y$ near $s_i$, $p(z\neq i|y)\le \epsilon$ for some small $\epsilon$; this is just a computational fact about fast-decaying tails from Gaussians.

If you believe the above, you'd believe that there's similar information carried in $p(x|y)$. For instance, for $y$ near $s_i$ since $X$ is likely sampled from $Z=i$'s Gaussian, $\CE{X}{Y}$ is near $s_i$ too.

In this sense $p(x|y)$ carries information about peaks, as its modes will be at values of $x,y$ simultaneously near $s_i$, and this is unique: as $x$ or $y$ moves further from $s_i$ its value decreases (as long as we're in the same neighborhood).

$p(y)$ more simply carries peak information, as observed by its local modes.

_(B) Let $p(y) = f(y)$ and $p(x) = g(x)$ be general probability distributions on $\R$ now. We can view $P_{Y|X}$ as a simple function of two variables, giving the conditional probability. For the Gaussian density $\phi$, $P_{Y|X}(x, y)=p(y|x)=p(N=y-x)=\phi(y-x)$. But we can lift it to a linear functional operator, $\mcP_{Y|X}$, with the property
$$
f(y)=\mcP_{Y|X}g=x\mapsto\int\d{x} g(x) P_{Y|X}(x, y)=\phi*g\,\,,
$$
where $*$ is the convolution operator. Why isn't the analogous $\mcP_{X|Y}$ just deconvolution?_

First of all, let's be clear about what deconvolution means.

We have this equality by the Fourier Inversion theorm, holding by $f\in L^1$, so
$$
f=\mcF^{-1}\mcF f=\mcF^{-1}\pa{\mcF g\mcF \phi}\,\,,
$$
wher we also use the Convolution Theorem. As $\pa{\mcF\phi}(\omega)=\exp\pa{\frac{-\omega^2\sigma^2}{2}}$ has no roots in $\C$, we can do some division, yielding
$$
\mcF^{-1}\frac{\mcF f}{\mcF\phi} = g\,\,,
$$
where $\mcG h=\mcF^{-1}\frac{\mcF h}{\mcF\phi}$ is deconvolution.

Now, Bayes' Theorem shows that
$$
P_{X|Y}(x, y)=\frac{g(x) P_{Y|X}(x, y)}{f(y)}\,\,,
$$
and lifting this to a transition kernel gives
$$
\mcP_{X|Y}f = \int\d{y}f(y)P_{X|Y}(x, y)=g\,\,,
$$
so probability still works consistently, but examining the transform,
$$
\mcP_{X|Y}h = x\mapsto \int\d{y}h(y)P_{X|Y}(x, y)=g\cdot\pa{\phi *(h/f)}\,\,,
$$
which is by inspection unequal to $\mcG$ (take any argument not equal to $f$).

An intuitive reason why these two might not be equal can be drawn from the causal nature of the system. The conditions for our Fourier Inversion theorem and Convolution Theorem are pretty flexible. As long as $f,\mcF$ are $L^1$ then there's a bijection $f\leftrightarrow g$. So if deconvolution was our mechanism $\mcP_{X|Y}$, we'd expect a very wide range of functions $g$ to perfectly recover functions $f$. Causal inference wouldn't make sense. There's no asymmetry or information loss from $X$ to $Y$.

There might be a deeper connection here based on the fact that $\phi$ is a [universal kernel](https://en.wikipedia.org/wiki/Kernel_embedding_of_distributions), but the Fourier explanation seemed good enough to me.

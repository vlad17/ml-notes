# Problem 10.8

_Prove Theorem 10.2, which states that for two full time graphs where $X_t^{j}$ is always influenced some previous version $X_{t-s}^{j}$ for some $s\ge 1$ the same between two graphs that are Markov equivalent and have acyclic summary graphs are equal._

Since the graphs are Markov equivalent, they share injunctions and skeletons. It suffices to show all edges are directed the same way. 

Since inflence requires total causal effect, there is some directed path from $X_{t-s}^j$ to $X_t^{j}$ in both graphs. However, these paths cannot include any variable $X_r^{k}$ with $k\neq j$, since that would create a summary graph cycle. Thus there is some minimal parent $X_{t-s'}^j$ with $s'\le s$ adjacent to $X_t^j$.

Due to the one direction time can move, and the fact that this undirected arrow exists in both graphs, it must also share direction in both graphs. Call it $e$.

Consider some edge $f$ in the first graph. $e$ must point into some $X_t^{j}$ in the first graph. An edge must exist in the other graph in the same place, albeit possibly in another direction. $f$ is either $e$, in which case the graphs agree, or forms an injunction in the first graph with $e$, in which case the graphs agree as well.

# Problem 10.9

_Consider an SCM over a multivariate time series where each variable is influenced by all past values of all components. The instantaneous effects form a DAG and the joint is Markovian and faithful with respect to the fulltime graph. To what extent can one identify the instananeous DAG structure from the distribution across time?_

In the iid sample model, we can recognize DAG structure under various scenarios, assuming restricted model classes (which ban conditional distributions or assign proposed structures which requires conditional distributions not fitting within the class a bad fit) or by testing independence (let's assume we're in a fully observable setting for now).

The question is to what degree can we identify the instantaneous DAG structure given a single time series realization?

One simple reduction is using mixing. The time series $X_1, \cdots X_t$ forms a Markov chain. If this chain is ergodic then in theory we can perform the same estimations that are used for independence testing in DAG identifiability using MCMC samples. We'd expect a reduced efficiency from sample autocorelation quantified from [Markov Chain CLT](https://en.wikipedia.org/wiki/Markov_chain_central_limit_theorem). But the idea is that you could just take a hit in effective sample size but still sample from the same chain.

That said, mixing isn't the end of the story. Under certain model assumptions, systems that otherwise mix poorly may have easily identifiable system dynamics. Arguments from [linear dynamical systems](https://arxiv.org/abs/1802.08334) might carry over to independence testing (it'd be a lot of work though).

_(E) Compute $\partial_X\CE{W}{\mathrm{do}(X)}$._

Note the almost sure equality from the equations:
$$
W = (5\alpha -3)X-2V + N\,\,,
$$
where $N$ is a linear zero-bias function of the node noises. The above holds for the passive observational case. Conditioning on $\mathrm{do}(X\gets 3)$, $N$ changes but otherwise the equality above still holds (and the causal path $V\rightarrow X$ vanishes).

Thus $\partial_X\CE{W}{\mathrm{do}(X)}=5\alpha-3$.

# 6.56

_For the SCM $Y\gets (X-4)^2+N_Y,Z\gets X^2+Y^2+N_Z$, with $X,N_Y,N_Z$ standard unit normal iid, what intervention on $X$ xor $Y$ yields the smallest expected $Z$?_

We compute the expectations under interventions.

$$
\CE{Z}{\mathrm{do}(X)} = \CE{X^2 + (X-4)^4 + 2N_Y(X-4)^2+N_Y^2+N_Z}{\mathrm{do}\pa{X}}\\ = X^2+(X-4)^4+1\,\,.
$$

On the other hand,
$$
\CE{Z}{\mathrm{do}(Y)} = \CE{X^2 + Y^2+N_Z}{\mathrm{do}\pa{Y}} = 1+Y^2+1\,\,.
$$

The latter is smaller with a minimizer $\mathrm{do}(Y=0)$ compared to any setting of $X$.

# Problem 6.57

_(A) Why does causal minimality imply structural minimality?_

We show this by contrapositive. Consider an SCM with graph $\mcG$ and suppose this graph isn't structurally minimal, admitting a subgraph $\mcG'$ which is the unique structurally minimal graph for the SCM with equivalent functional specification of the SCM. It suffices to only examine a single node in $\mcG'$ whose parents differ from those in $\mcG$. By Proposition 6.49, the SCM on $\mcG$ and the SCM on $\mcG$ with one of the nodes with its parents restrcited to the strict subset of parents that the node has in $\mcG'$ are counterfactually equivalent, implying no causal minimality for $\mcG$.

_(B) What's a structurally minimal SCM that's not causally minimal?_

Consider the following SCM:
$$
A\gets N(0,1)\\
B\gets A, C\gets -A\\
D\gets B + C + N(0,1)
$$

By the functional form of its assignments, the graph with arrows $A\rightarrow B,C$ and $B,C\rightarrow D$ is structurally minimal.

However, since $B,C$ are determined completely by $A$ (they have no noise in this degenerate example), the same joint distribution is Markovian with respect to the graph over $A,B,C,D$ with just the two arrows $A\rightarrow B,C$, and thus the original larger graph is not causally minimal.



# Problem 6.58

_By Proposition 6.36, any non-causally-minimal SCM for a strictly postive, continuous density $P$ has an inactive edge $Y\rightarrow X$ where $X\independent Y|\mathrm{Pa}(X)\setminus Y$. Show removing this edge leads to an interventionally equivalent SCM._.

Per Proposition 6.48, it suffices to show that for any node $A$ with a parent $B$ in either of the graphs, the equality $p(A|\mathrm{do}(B), \mathrm{Pa}(A)\setminus B)=p'(A|\mathrm{do}(B), \mathrm{Pa}(A)\setminus B)$ holds. This is directly true for all pairs of nodes since the amputated graph is identical except at one edge.

The only case worth examining is $p(X|\mathrm{do}(Y), \mathrm{Pa}(X)\setminus Y)$ vs $p'(X|\mathrm{do}(Y), \mathrm{Pa}(X)\setminus Y)$. But in both cases this is equal to $p(X|\mathrm{Pa}(X)\setminus Y)$, due to separation between $X$ and $Y$ given the parents.

# Problem 6.59

_Consider the two SCMs

Y\gets X+\Uniform[-1,1]\\

Find two functions such that the SCMs agree observationally and on all single node interventions, but disagree on simultaneous interventions on several nodes, showing that the assumption of Proposition 6.48, which requires almost sure inequality for a full-support soft intervention, is load-bearing._

Consider $f_1(X, Y)=\boldsymbol 1\ca{XY=1},f_2=0$. This example also makes the characterization theorem of Proposition 6.48 pretty weak. It doesn't apply to discrete variables, for instance.

# Problem 6.60

_Prove that whenever all absolute eigenvalues of a square matrix $B$ are strictly smaller than 1 then $I-B$ is invertible._

Fix an eigen pair $(\vv, \lambda)$ of $B$. Notice $(I-B)\vv=(1-\lambda)\vv$, so $I-B$ has the same eigenspace but eigenvalues $\abs{1-\lambda_i}>0$, though they could be complex. Nonetheless, they're not zero and thus the matrix is invertible.

# Problem 6.61

_Consider the cyclic linear SCM $X\gets BX+N$ for a random vector $X$ and independent vector noise $N$. Viewing the SCM as an iterative process with fixed noise, it can be viewed as the limit of $X_t=BX_{t-1}+N$ for $t\rightarrow\infty$. What is the limit, assuming $\rho(B)<1$?_

Unrolling the recursion, we find that $X_\infty=\sum_{t=0}^\infty B^tN$. The limit $\sum_tB^t=(I-B)^{-1}$ exists (its proof is identical to that of geometric sums, but done with matrices). The only technicality is that the limit of terms with a diagonal eigenvalue matrix exponentiated to $t$ vanishes, which holds by $\rho(B)<1$.

# Problem 6.63

_Suppose $Z$ is a valid adjustment set for the causal effect from $X$ to $Y$ and that $(Y, X, Z)$ is a zero mean Gaussian with
$$
\CE{Y}{X,Z}=\valpha^\top X+\vbeta^\top Z\,\,.
$$
Then show that this regression recovers coefficients $\valpha$ for the causal effect with this covariate adjustment (regressing $Y$ on $(X, Z)$)._

$$
\begin{align}
\nabla_X\CE{Y}{\mathrm{do}(X)}&=\nabla_X\CE{\CE{Y}{\mathrm{do}(X), Z}}{\mathrm{do}(X)}\\
&=\nabla_X\CE{\CE{Y}{X, Z}}{\mathrm{do}(X)}\\
&=\nabla_X\CE{\valpha^\top X+\vbeta^\top Z}{\mathrm{do}(X)}\\
&=\nabla_X\pa{\valpha^\top X+\CE{\vbeta^\top Z}{\mathrm{do}(X)}}\\
&=\valpha+\CE{\nabla_X\vbeta^\top Z}{\mathrm{do}(X)}\\
&=\valpha
\end{align}
$$
following by the tower property, then action/observation exchange, the given conditional expectation, linearity, and then regularity (smoothness of the joint Gaussian measure $(X, Z)$ and its integrand $\vbeta^\top Z$ allows interchange).

# Problem 6.64

_Prove parent adjustment and the backdoor criterion for a DAG SCM with $Y\not\in P = \mathrm{Pa}(X)$._

For both parents and blocked backdoors, any paths from $X$ to $Y$ must go through $X$'s children. Create a virtual intervention node $I_X$ which, if it's not $\bot$, which is its usual value, it sets $X$ to whatever it's set to. It's clear this is equivalent to intervention on $X$, but now it lets us show that $I_X$ is blocked by $X$ onto any path to $Y$, and thus we may exchange interventions $I_X$ with conditioning on $X$ since $I_X$ is completely independent of $Y$ given $X$ and its parents (or backdoor blockers).

_Prove frontdoor adjustment_.

$$

by marginalization,

$$

$$
by action/observation interchange, as there's no backdoor paths from $X$ to $Z$.
